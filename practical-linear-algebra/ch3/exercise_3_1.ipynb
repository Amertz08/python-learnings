```python
import numpy as np
```

```python
scalars = [1, 2, -3]
vectors = [np.array([4,5,1]), np.array([-4,0,-4]), np.array([1,3,2])]

output = np.zeros(len(scalars))

for i, v in enumerate(vectors):
    output += v*scalars[i]
output
```

```python
# double check with using a dot product matrix multiplication
np.array(scalars) @ np.array(vectors)
```
