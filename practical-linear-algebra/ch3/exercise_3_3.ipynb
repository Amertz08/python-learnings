```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objs as go
```

```python
vector_set = np.array([1,3])
vector_set
```

```python
n_count = 200
first_scalars = np.random.default_rng().uniform(-4,4,n_count)
```

```python
points = []
for val in first_scalars:
    points.append(val * vector_set)
```

```python
plt.scatter(x=[p[0] for p in points], y=[p[1] for p in points])
plt.arrow(x=0, y=0, dx=1, dy=3, head_width=0.1, head_length=0.1, length_includes_head=True)
plt.axis('equal')
plt.show()
```

Our subspace is defined by the vector [1,3] and so any scalar multiplied by that vector creates a point in the subspace. Subspaces combine any linear weighted combination of the spanning vectors.

```python
v1 = np.array([3,5,1])
v2 = np.array([0,2,2])
```

```python
second_scalars = np.random.default_rng().uniform(-4,4,(n_count, 2))

second_points = np.zeros((n_count,3))
for i, _ in enumerate(second_scalars):
    second_points[i,:] = v1*second_scalars[i, 0] + v2*second_scalars[i, 1]
```

```python
fig = go.Figure(data=[go.Scatter3d(x=second_points[:,0], y=second_points[:,1], z=second_points[:,2],mode='markers')],)
fig.show()
```
