```python
import math

import numpy as np
```

```python
a = np.array([1, 2])
```

```python
np.dot(a, a)
```

```python
# the squared norm is == to the dot product of a vector onto itself
np.linalg.norm(a) ** 2
```

```python
# this would also mean that the sqrt of a vector dot product onto itself is the norm
math.sqrt(np.dot(a, a))
```

```python
np.linalg.norm(a)
```
