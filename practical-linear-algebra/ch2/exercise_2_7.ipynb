```python
import numpy as np
```

```python
a = np.array([1, 2])
b = np.array([3, 4])
```

```python
# dot product is commutative. Multiplication is commutative a x b == b x a and the dot product is just the sum of every element multiplied by the other element, then the whole operation is commutative.
np.dot(a, b) == np.dot(b, a)
```
