```python
import numpy as np
```

```python
def unit_vector(vector: np.ndarray) -> np.ndarray:
    return vector / np.linalg.norm(vector)
```

```python
# we know that the unit vector has a magnitude of 1 and so we simply need to multiply
# the unit vector by the desired magnitude vector to get a vector with that very magnitude
def mag_vector(vector: np.ndarray, magnitude: float) -> np.ndarray:
    return unit_vector(vector) * magnitude
```

```python
v = mag_vector(np.array([1, 2, 3]), 2)
v
```

```python
np.linalg.norm(v)
```
