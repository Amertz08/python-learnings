```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = np.array([1, 2])
w = np.array([4,-6])
v_plus_w = v + w
```

```python
plt.figure(figsize=(10,10))
```

```python
a1 = plt.arrow(x=0,y=0, dx=v[0], dy=v[1], head_width=0.3, width=0.1, color='k', length_includes_head=True)
a2 = plt.arrow(x=v[0], y=v[1], dx=w[0], dy=w[1], head_width=0.3, width=0.1, color='r', length_includes_head=True)
a3 = plt.arrow(x=0, y=0, dx=v_plus_w[0], dy=v_plus_w[1], head_width=0.3, width=0.1, color='b', length_includes_head=True)
plt.legend([a1,a2,a3],['v','w','v+w'])
```

```python
v_minus_w = v - w
```

```python
plt.figure(figsize=(10,10))

a1 = plt.arrow(x=0,y=0, dx=v[0], dy=v[1], head_width=0.3, width=0.1, color='k', length_includes_head=True)
a2 = plt.arrow(x=0, y=0, dx=w[0], dy=w[1], head_width=0.3, width=0.1, color='r', length_includes_head=True)
a3 = plt.arrow(x=w[0], y=w[1], dx=v_minus_w[0], dy=v_minus_w[1], head_width=0.3, width=0.1, color='b', length_includes_head=True)
```
