```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([2, 10])
b = np.array([3, 2])
beta = np.dot(a.T, b) / np.dot(a.T, a)
beta
```

```python
b_onto_a = beta * a
c = b - b_onto_a
b_onto_a, c
```

```python
plt.figure(figsize=(8, 8))
```

```python
a1 = plt.arrow(x=0,y=0, dx=a[0], dy=a[1], color='r', width=0.1, length_includes_head=True)
a2 = plt.arrow(x=0,y=0, dx=b[0], dy=b[1], color='b', width=0.1, length_includes_head=True)
a3 = plt.arrow(x=0, y=0, dx=b_onto_a[0], dy=b_onto_a[1], color='g', width=0.05, length_includes_head=True)
a4 = plt.arrow(x=b_onto_a[0], y=b_onto_a[1], dx=c[0], dy=c[1], color='y', width=0.05, length_includes_head=True)
plt.axis("equal")
```
