```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kernel = np.array([0, 0.1, 0.3, 0.8, 1, 0.8, 0.3, 0.1, 0])
kernel_normed = kernel / kernel.sum()
kernel_normed.sum()
```

```python
counts = 100
x_axis = np.arange(counts)
data = np.random.default_rng().normal(0, 2, size=counts)
data[:10]
```

```python
def apply_kernel(k, signal):
    ans = np.zeros(len(signal))
    for i in range(len(k), len(signal)):
        ans[i] = np.dot(k, signal[i-len(k):i])
    return ans
```

```python
answer = apply_kernel(kernel_normed, data)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x_axis, data)
ax.plot(x_axis, ans)
fig.legend(["Signal", "Smoothed"])
```

```python
# exercise 4.7 high pass filter
high_pass_kernel = np.array([0, 0.1, 0.3, 0.8, -1, 0.8, 0.3, 0.1, 0])
high_pass_normed = high_pass_kernel - high_pass_kernel.mean()
```

```python
hp_ans = apply_kernel(high_pass_normed, data)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x_axis, data)
ax.plot(x_axis, hp_ans)
fig.legend(["Signal", "Smoothed"])
```
