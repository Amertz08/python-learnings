```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kernel = np.array([0, 0.1, 0.3, 0.8, 1, 0.8, 0.3, 0.1, 0])
kernel_normed = kernel / kernel.sum()
kernel_normed.sum()
```

```python
counts = 100
x_axis = np.arange(counts)
data = np.random.default_rng().normal(0, 2, size=counts)
data[:10]
```

```python
chunk_size = len(kernel_normed)
ans = np.zeros(len(x_axis))
for i in range(chunk_size, len(data)):
    ans[i] = np.dot(kernel_normed, data[i-chunk_size:i])
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x_axis, data)
ax.plot(x_axis, ans)
fig.legend(["Signal", "Smoothed"])
```
