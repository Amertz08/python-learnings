```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kernels = np.array([[-1,1], [1, -1], [-1, 2]])
```

```python
a = [-1] * 10 + [1.5] * 5 + [-1] * 10
signal = np.array(a)
signal
```

```python
def edge_detect(k, sig):
    ans = np.zeros(len(sig))
    for i in range(1, len(sig)-1):
        segment = np.array([sig[i-1], sig[i]])
        ans[i] = np.dot(k, segment)
    return ans
```

```python
answers = []
for k in kernels:
    answers.append(edge_detect(k, signal))
```

```python
fig, axes = plt.subplots(nrows=len(kernels), figsize=(10, 10))
x_axis = np.arange(len(signal))
for i, ans in enumerate(answers):
    axes[i].scatter(x_axis, ans)
    axes[i].scatter(x_axis, signal)
    axes[i].set_title(f"Kernel: {k}")
# ax1 = fig.add_subplot()
# ax1.scatter(x=x_axis, y=answer)
# ax1.scatter(x=x_axis, y=signal)
```
