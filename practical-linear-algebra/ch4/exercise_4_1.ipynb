```python
import numpy as np
```

```python
v1 = np.array([1, 2])
v2 = np.array([3, 4])
```

```python
def find_vals(a: np.array, b: np.array) -> tuple[np.float64, np.float64]:
    a_mean = np.mean(a)
    b_mean = np.mean(b)

    x = np.sqrt(np.dot((a - a_mean).T, a - a_mean))
    y = np.sqrt(np.dot((b - b_mean).T, b - b_mean))

    coeff = np.dot((a - a_mean).T, (b - b_mean)) /  (x*y)

    cosin = np.dot(a.T, b) / (np.sqrt(np.dot(a.T, a)) * np.sqrt(np.dot(b.T, b)))

    return coeff, cosin
```

```python
# values are different because we did not mean center the input
x, y = find_vals(v1, v2)
x, y
```

```python
# sanity check
np.corrcoef(v1, v2)
```

```python
# sanity check
np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
# if we mean center both before calling then the values are the same
find_vals(v1-np.mean(v1), v2-np.mean(v2))
```
