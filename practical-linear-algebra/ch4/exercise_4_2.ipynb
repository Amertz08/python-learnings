```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vals = np.arange(4,dtype=np.float64)
offsets = np.arange(-50,50)
results = np.zeros((len(offsets), 2))
```

```python
def find_vals(a: np.array, b: np.array) -> tuple[np.float64, np.float64]:
    a_mean = np.mean(a)
    b_mean = np.mean(b)

    x = np.sqrt(np.dot((a - a_mean).T, a - a_mean))
    y = np.sqrt(np.dot((b - b_mean).T, b - b_mean))

    coeff = np.dot((a - a_mean).T, (b - b_mean)) /  (x*y)

    cosin = np.dot(a.T, b) / (np.sqrt(np.dot(a.T, a)) * np.sqrt(np.dot(b.T, b)))

    return coeff, cosin
```

```python
for i, offset in enumerate(offsets):
    results[i,:] = find_vals(vals, vals+offset)
```

```python
plt.figure(figsize=(8,4))
plt.plot(offsets, results)
plt.legend(["Pearson Coefficient", "Cosine Similarity"])
```
