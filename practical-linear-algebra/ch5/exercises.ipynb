# Exercise 5-5

Matrix addition and scalar multiplication obey commutivity and distributivity. Prove via Python

```
sigma*(A + B) = sigma*A + sigma*B = A*sigma + B*sigma
```

````python
A = np.round(np.random.default_rng().random((3,4)), 2)
B = np.round(np.random.default_rng().random((3,4)), 2)
sigma = 4

print(sigma * (A + B) == sigma * A + sigma * B)
print(sigma * A + sigma * B == A * sigma + B * sigma)
````

````python
matrices = [A, B, np.round(sigma*(A + B), 2), np.round(sigma*A + sigma*B, 2)]
fig, ax = plt.subplots(nrows=len(matrices), figsize=(10,10))

for i, matrix in enumerate(matrices):
    ax[i].matshow(matrix)

    for x in range(matrix.shape[0]):
        for y in range(matrix.shape[1]):
            val = matrix[x, y]

            ax[i].text(y, x, str(val), ha="center", va="center")
````

````python
symmetric = np.array([[1.0, 2.0], [2.0, 1.0]])
plt.matshow(symmetric)
````

````python
symmetric.T == symmetric
````

# Exercise 5-9

Additive method for creating a symmetric matrix from a nonsymmetric square matrix. Average the matrix with it's transpose.

````python
non_symmetric = np.array([[1.1, 2.0], [2.1, 1.0]])
````

````python
avg_matrix = (non_symmetric + non_symmetric.T) / 2
avg_matrix
````

````python
avg_matrix.T == avg_matrix
````

# Exercise 5-11
- Premultiplying by a diagonal matrix scales the rows of the right matrix by the corresponding diagonal elements
- Postmultiplying by a diagonal matrix scales the columns of the left matrix by the corresponding diagonal elements

````python
ones = np.ones((4,4))
diagonal = np.diag([1,4,9,16])
diagonal_sqrt = np.sqrt(diagonal)
````

````python
# premultiply and thus scale the rows
diagonal @ ones
````

````python
# post multiply and scale the columns
ones @ diagonal
````

````python
diagonal_sqrt @ ones @ diagonal_sqrt
````
