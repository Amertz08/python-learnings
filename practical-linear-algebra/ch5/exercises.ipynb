Matrix addition and scalar multiplication obey commutivity and distributivity. Prove via Python

```
sigma*(A + B) = sigma*A + sigma*B = A*sigma + B*sigma
```

````python
A = np.random.default_rng().random((3,4))
B = np.random.default_rng().random((3,4))
sigma = 4

print(sigma * (A + B) == sigma * A + sigma * B)
print(sigma * A + sigma * B == A * sigma + B * sigma)
````
